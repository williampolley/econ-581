## Exponential and Logarithmic Functions
Exponential and logarithmic functions are fundamental to all facets of economics and finance. In this section, we will explore these functions so that you have a better undertanding of their many uses.

Consider the following:

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def econplot(x,series,x_min,x_max,y_min,y_max,chart_title,chart_x,chart_y):
    fig = plt.figure()
    fig.set_figheight(6)
    fig.set_figwidth(6)
    ax = fig.add_axes([0,0,1,1])
    ax.axis([x_min,x_max,y_min,y_max])
    if np.ndim(series)==1:
        ax.plot(x,series)
    else:
        for y in series:
            ax.plot(x,y)
    ax.set_xlabel(chart_x,size=16)     
    ax.set_ylabel(chart_y,size=16)    
    ax.set_title(chart_title,size=24)
    ax.tick_params(which='both',labelsize=16)
    return fig,ax
```

```python
t=np.linspace(-2,5,100)
def y(t):
    return 2**t
fig,ax=econplot(t,y(t),-2,5,0,35,'2^t','t','2^t')
plt.show()
```

```python
t=np.linspace(-2,5,100)
def y(t):
    return 3**t
fig,ax=econplot(t,y(t),-2,5,0,55,'3^t','t','3^t')
plt.show()
```

Now consider this:

The shape of these two functions is the same, only the scale has changed. In a sense, every exponential function is the same shape. For this reason, it is useful to have a common base whenever dealing with exponential functions. That common base is the number "$e$". You might think the "$e$" stands for exponential, but actually it honors the great Swiss mathematician Leonhard Euler (pronounced "oiler"). Euler's number is an irrational number, meaning its decimal representation neither terminates nor repeats.

```python
t=np.linspace(-2,5,100)
fig,ax=econplot(t,np.exp(t),-2,5,0,55,'e^t','t','e^t')
plt.show()
```

To use the exponential function, call the NumPy function `np.exp(t)` as in the example above. Note also that one of the most interesting mathematical properties of the exponential function is that it is its own derivative. We can approximate this numerically.

```python
from scipy.misc import derivative
def d_exp(x):
    return derivative(np.exp,x,dx=10**(-10))
y=np.array([d_exp(t),np.exp(t)])
fig,ax=econplot(t,y,-2,5,0,55,'e^t and d(e^t)/dt','t','')
```

You only see one curve because they are identical.

```python
diff=d_exp(t)-np.exp(t)
print(diff)
```

Remember that we used a $dx$ of $10^{-10}$ in the numerical derivative function, so this accuracy is pretty good.

The function $ae^{rt}$ is the function that describes continous compound growth. That is, the function above is the value of "$a$" dollars earning continously compounded interest for "$t$" periods at a periodic interest rate "$r$." Can you interpret the rate of growth of this function?

Now let us consider logarithms. Logarithms are exponents. Try this on your calculator: log 100. If your calculator is like most, the "log" key is a base 10 logarithm. That is, the logarithm (base 10) of 100 is the exponent to which you would raise the number 10 in order to equal 100. The answer is, of course, 2. In class, I will elaborate a bit more on this point.

When dealing with logarithms, it is important to specify the base. Now as we have found it useful to use the number $e$ as the base for thinking about exponential functions, it is natural that $e$ is a useful base for logarithms (which are exponents... repeat after me:  logarithms are exponents!) The logarithm using $e$ as its base has a special name:  the natural logarithm. The natural logarithm function is the inverse of the exponential function.

In economics, we really have little use for logarithms in any base other than $e$. Therefore, in economics we rarely say "natural" logarithm. We just say logarithm (or $log$). In some textbooks (and on most calculators) the natural log is written as $ln$. In economics, we often use both $log$ or $ln$ to mean the natural logarithm. Personally, I use $ln$, but it is not unusual to see $log$ in some papers or books.

NumPy uses `np.log()` to represent the natural logarithm.

```python
print(np.log(np.exp(1)))
print(np.log(100))
x=np.linspace(.1,4,200)
fig,ax=econplot(x,np.log(x),0,4,-2.5,2,'ln(x)','x','')
```

Because logarithms are exponents, they obey the laws of exponents. We will discuss this and some other aspects of logarithms in class.
