### Some history and context
Here we start at the very beginning especially for those who have limited experience with programming languages. Python is a "high level" language, which means that it is closer to human language as opposed to the binary language of computers. This makes Python relatively easier to learn than, for example, the C programming language. Trust me, I learned C in college. Python is a breeze by comparison. However, high level languages are interpreted rather than compiled. C is a compiled language. When you write code in C, it needs to be compiled into a binary executable file. The executable file runs very fast as it is in the computer's native language, but the process of compiling and re-compiling the code after any minor change can be long and tedious. High level languages like Python run through an interpreter at run-time and do not need to be compiled. The resulting code runs more slowly, but the debugging process is a lot faster.

When desktop computers were slow by today's standards and had much lower memory capacity, scientific applications needed to use compiled languages like C or Fortran to be able to accomplish a task in an acceptable amount of time. Interpreted languages like BASIC were easy to learn, but too slow to be useful in cutting edge science. Python has been available as a language since the early 1990s, but did not gain wide acceptance until more recently when computing power reached a point where even a relatively slow language could still be fast enough to solve big problems. Of course, in applications where speed matters, programmers utilize C libraries and compile the code before deployment into production. Those considerations are mainly the realm of computer scientists. Economists and data scientists generally do not need to get into those details. Our work can now largely be done by high level languages like Python at a speed that is acceptable for our tasks. For example, if I need to invert a large matrix, it might take Python one second to do it. Compiled code could do it in a fraction of a second. If I need to do it a million times, the difference might be significant, but if I only need to do it once, the ease of coding in an interpreted language is worth the slower speed. As economists would say, it's all about the trade-offs.

## Functions
As a prerequisite for this section, you should read the page on [functions](https://python-programming.quantecon.org/functions.html) in Sargent and Stachurski. We will go over it in class as well.

Once you are comfortable with the material presented there, take a look at the following example. Something that we do a lot in mathematical economics is to take the derivative of a function. We will then often want to graph the function with a tangent line, the slope of which is the derivative of the function at that point. This example comes from the [Math and Physics Problems Wiki](https://math-physics-problems.wikia.org/wiki/Graphing_Tangent_Lines_with_Python) with only a slight modification.

```python
def deriv(f,x):
    h = 0.000000001
    return (f(x+h) - f(x))/h


def tangent_line(f,x,x0):
    y = f(x) 
    y0 = f(x0)
    return deriv(f,x0) * (x - x0) + y0
```

Running the code produces no output, but it does define the functions `deriv(f,x)` and `tangent_line(f,x,x0)`. Notice how we call `deriv` in the return statement of `tangent_line`. It is important that the code that defines `deriv` comes before the code that defines `tangent_line` for that reason. You can see that deriv is simply a numerical approximation of the derivative of a function at a point. Note that we are simply using the calculus definition of the derivative. Technically, the derivative of a function at a point is the limit of `(f(x+h) - f(x))/h` as $h$ goes to zero. Numerical computation on computers cannot really handle the idea of a limit very well. We can make the step size as small as the machine precision will allow and approximate the derivative.

We should note that there are other, better, ways to implement a derivative function in Python. Numpy has a function to take a derivative of a polynomial. (It's just an implementation of the power rule in calculus--very simple.) Other packages such as scipy and sympi implement a derivative function as well. However, for our purposes this function does the job. The purpose of this is simply to generate a tangent line that we can draw on a graph. The code is very simple for the purposes of illustration.

Now let's combine this with what we learned earlier about plotting. We will define yet another function to handle the plotting. Yes, we can do that. Python is a versatile language. Here is the function that I will call `econplot`. We will use this function repeatedly as it automates some of the tasks associated with creating a chart that we do not want to repeat every single time.

```python
def econplot(x,series,x_min,x_max,y_min,y_max,chart_title,chart_x,chart_y):
    fig = plt.figure()
    fig.set_figheight(6)
    fig.set_figwidth(6)
    ax = fig.add_axes([0,0,1,1])
    ax.axis([x_min,x_max,y_min,y_max])
    if np.ndim(series)==1:
        ax.plot(x,series)
    else:
        for y in series:
            ax.plot(x,y)
    ax.set_xlabel(chart_x,size=16)     
    ax.set_ylabel(chart_y,size=16)    
    ax.set_title(chart_title,size=24)
    ax.tick_params(which='both',labelsize=16)
    return fig,ax
```

We have a few differences between this and the plotting code we looked at earlier. Let's examine what has changed. First of all, `fig` and `ax` are called into existence separately. This just makes it easier to see what we are doing with each. The methods `set_figheight` and `set_figwidth` are used to change the dimensions of the container that the figure will occupy. The axes are then set up with the dimensions and labels as desired. Notice that we are putting things like the axes bounds, labels, and title into the list of parameters for the function. This allows us to use the same function to potentially create different charts with different series and associated labels and titles. The function returns the figure and axes `(fig,ax)` which define the chart.

Note that when we call the function, we pass "series" as a parameter. Then when the code executes, we first check to see if there is just one series or if it a list of several series. If it is just one, we plot "series." If there is a list, we loop through series and plot each "y" in the series against x. Why do it this way? Because now we have the flexibility to plot one or more curves on the same figure with a single function call.

Let's see how it works.

```python
x=np.linspace(0,100,101)  #define a domain for x
def y1(x):
    return x**2  #define a function y1 the square of x
t=tangent_line(y1,x,70)  #get the equation of a line tangent to y1 at x=70
y=np.array([y1(x),t])  #create an array containing the function and the tangent line
fig,ax=econplot(x,y,0,100,0,10000,'x squared','x','y') #call the econplot function which will return a fig,ax pair
plt.show()  #show the plot of the fig,ax pair that was returned
```

Nice! Now that we have defined simple functions to do the hard work, we can create a graph of a function and it's tangent with just a few very simple and easy to understand lines of code! Suppose we were writing a paper or presentation where we had to create a lot of these. We make our job very easy. ## Polynomials
Now let us consider how to plot a polynomial function of the form $$a_0 + a_1 x + a_2 x^2 +...+ a_n x^n.$$ In the previous section, we were introduced to the plotting of a quadratic function and its derivative (which is linear). As we saw there, we often write equations in slope-intercept form because that makes them very easy to plot in the coordinate plane. Consider a line with a slope of $2$ and a $y$-intercept of $10$.  We might write in mathematical notation, $y = f(x) = 2x + 10$. Using our `econplot` function we have already defined, we can plot this function with the following code.

```python
x=np.linspace(-10,10,100)
def y(x):
    return 2*x+10
fig,ax=econplot(x,y(x),-20,20,-10,30,'A Linear Function','x','y')
ax.grid()
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_label_coords(1,.2)
ax.set_ylabel('y',size=16,rotation=0)
ax.yaxis.set_label_coords(.4,.95)
plt.show()
```

Note the additional methods used to move the spines so that the axes' tick marks eminate from the origin. This helps the viewer to see the quadrants of the coordinate plane more easily.

Next, a quadratic... $x^2 + 4x - 16$

```python
x=np.linspace(-10,10,100)
def y(x):
    return x**2+4*x-16
fig,ax=econplot(x,y(x),-10,10,-20,60,'A Quadratic Function','x','y')
ax.grid()
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_label_coords(1,.2)
ax.set_ylabel('y',size=16,rotation=0)
ax.yaxis.set_label_coords(.4,.95)
plt.show()
```

And a cubic...  $x^3 - 4x^2 + x - 3$

```python
x=np.linspace(-10,10,100)
def y(x):
    return x**3-4*x**2+x-3
fig,ax=econplot(x,y(x),-2,5,-20,8,'A Cubic Function','x','y')
ax.grid()
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_label_coords(1,.65)
ax.set_ylabel('y',size=16,rotation=0)
ax.yaxis.set_label_coords(.22,.95)
plt.show()
```

### Solving polynomial equations
Now that you have seen how to plot a polynomial function, the next step is to solve a polynomial equation. Recall that a polynomial function becomes a polynomial equation when we set it equal to zero. Graphically, the solution(s) to this equation are the "zeros" of the function, that is, where the graph of the function crosses the $x$-axis. For example, to solve the equation $y = 2x + 10$ we have the following code.

```python
np.roots([2,10])
```

The function `np.roots()` takes as its argument a list of the coefficients of the polynomial equation and returns an array containing the roots of the equation (or zeros of the function). In this case, the solution to the linear equation $2x + 10 = 0$ is $x = -5$. Similarly, we can solve for the zeros of the quadratic and cubic functions illustrated above.

```python
print(np.roots([1,4,-16]))
print(np.roots([1,-4,1,-3]))
```

Note that if we put multiple statements like this into a single cell, they will not all display unless we call the `print()` function. Note also that in the case of the cubic, there is only one real root for that equation (the one place where the curve crosses the $x$-axis). If there are any complex roots, then all of the roots (including any real roots) will be expressed as a complex number. Python uses `j` rather than the more common `i` for the imaginary number (square root of -1).

We can use the functions `np.isreal()` and `np.real()` to identify the real roots if those are the only ones in which we are interested (which is usually the case in economics).

```python
for i in np.roots([1,-4,1,-3]):
    if np.isreal(i):
        print(np.real(i))
```

### Solving systems of linear equations
Often in economics we want to solve systems of equations. If those equations are linear, this is an basically a problem of inverting a matrix. As long as our computing capacity is sufficient, the problem is straightforward. If the equations are non-linear, then it is another matter. Systems of non-linear equations can be extremely difficult to solve. Many books and articles have been devoted to various methods. Often the best route is to do a linear approximation to the non-linear functions that will have sufficiently good accuracy locally. If we have time in this course, we may look at these methods. However, for now we will look at solving systems of linear equations. This is just a quick introduction as we will dive more deeply into the mathematics in the next section.

Consider the following system:

$$x + y = 10$$

$$3x - 4y = 15$$

```python
a = np.array([[1,1],[3,-4]])
b = np.array([10,15])
print(np.linalg.solve(a,b))
```

[Documentation](https://numpy.org/doc/stable/reference/generated/numpy.linalg.solve.html) for this function explains the syntax in detail. In short, we need to define two arrays and solve the matrix algebra problem $aX = b$. (More on this in the next section.) The matrix a contains the coefficients on the left-hand side of the equations. The matrix $b$ is a vector containing the right-hand side of each equation. The function `np.linalg.solve()` returns an array containing the solved values for each variable in the order that you listed them in matrix $a$ (in this case, $[x,y]$).

Of course, one of the first things an economic student learns to do is solve a simple system of linear equations to obtain the equilibrium of supply and demand. We can do this for a simple example. Consider the system:

$$p = 2q + 10$$

$$p = 30 - q$$
where the equations represent the supply and demand relationships, respectively.

```python
a = np.array([[1,-2],[1,1]])
b = np.array([10,30])
print(np.linalg.solve(a,b))
```

So the solution is $p = 23.33$ and $q = 6.67$. We can plot this very easily using our `econplot` function.

```python
q=np.linspace(0,15,16)
def ps(q):
    return 2*q+10
def pd(q):
    return 30-q
y=np.array([ps(q),pd(q)])
fig,ax=econplot(q,y,0,15,0,40,'Supply and Demand','quantity','price')
ax.grid()
plt.show()
```

Conceptually, we think of quantity supplied or quantity demanded as being a function of price. The reader who is paying close attention will notice that in this code we are actually dealing with the inverse demand and inverse supply functions (price as a function of quantity). This is, of course, because we are following the usual convention of plotting quantity on the horizontal axis. Economists have been drawing their supply and demand diagrams backwards like this ever since Alfred Marshall. But that is a topic for another day.
